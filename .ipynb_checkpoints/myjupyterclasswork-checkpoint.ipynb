# while loops
A while loop is a fundamental programming construct in python that allows you to repeatedly execute a block of code as long as a specified condition is true. it's used when you want to perform a task repeatedly until a certain condition is met or until a specific number of iteration have been completed...note that you must terminate it yourself unless it will keep running

```python
#every while block of code start with the while statememt
count = 0
while count <= 5:
    print(count)
    count = count + 1
```

# user input loop
repeatedly asking the user for input until they enter 'quit'

```python
while True:
    user_input = input("Enter 'Quit' to exit: ")
    if user_input == "quit":
        print("Thank you for coming")
        break
    else:
        print("you entered:", user_input)
```

```python
def square(x):
    return x ** 2
```

```python
#Giving class an attribute
class Item:
    pay_rate = 0.8
    def __init__(self,name: str, price:float, quantity=0):
        assert price >=0, f'price{price} is not greater than or equal to zero'
        assert price >=0, f'quantity{quantity} is not greater than or equal to zero'
        self.name = name
        self.price = price
        self.quantity = quantity
    def calculate_total_price(self):
        return self.price * self.quantity
    def apply_discount(self):
            self.price = self.price * Item.pay_rate
        
```

```python
item1 = Item('phone', 1000, 5)
item2 = Item ('laptop', 3000, 3)
```

```python
item1.apply_discount()
```

```python
item1.price
```

```python
#Giving class an attribute
class Item:
    pay_rate = 0.8
    all = []
    def __init__(self,name: str, price:float, quantity=0):
        assert price >=0, f'price{price} is not greater than or equal to zero'
        assert price >=0, f'quantity{quantity} is not greater than or equal to zero'
        self.name = name
        self.price = price
        self.quantity = quantity
        Item.all.append(self)
    def calculate_total_price(self):
        return self.price * self.quantity
    def apply_discount(self):
            self.price = self.price * Item.pay_rate
        
```

```python
item1 = Item('phone', 1000, 5)
item2 = Item ('laptop', 3000, 3)
item3 = Item ('cable', 10, 5)
item4 = Item ('mouse', 50, 7)
item5 = Item ('keyboard', 75, 5)
```

# INHERITANCE
INHERITANCE CONCEPT IN OOP IS A PROCESS WHERE THE CHILD CLASS INHERIT ATTRIBUTES AND METHODS FROM THE PARENT CLASS. THE CHILD CLASS IS ALSO CALLED THE SUBCLASS or derived class and the existing class is called the superclass or base class.

```python
#define a superclass(base class)
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return "This Animal can make a sound" #tis method will be overridden
```

```python
#define a subclass
class Dog(Animal):
    pass
```

```python
class Cat(Animal):
    pass
```

```python
import numpy as np
```

```python
def filling_missing_data(column):
    if column.dtype == np.dtype("int64"):
        return column.fillna(column.mean())
        
    if column.dtype ==np.dtype("float"):
        return column.fillna(column.mean())
        
    if column.dtype ==np.dtype("object"):
        return column.fillna(column.mode()[0])
    
    #this is how to fill the missing values of int,object and float values in a dataset automatically.
    #mean and median is used to fill integer or floats,mode is used to fill texts or objects.
```

```python
for column in df:
    df[column] = filling_missing_data(df[column])
    
```

```python
df.isnull().sum()
```

```python
 #np.dtype()
```

# HISTOGRAM

```python
#HISTOGRAM(USED WHEN COMPARING RELATIONSHIP BETWEEN TWO NUMERICAL VARIEABLES)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,92,129]
blood_sugar_women = [67,98,89,120,135,150,84,69,89,79,120,112,100]

plt.hist(blood_sugar_men, bins=[80,100,125,150], rwidth=0.95, color="g");
```

# two categorical variables
side by side(or stacked) bar plots are useful for visualizing the relationship between two categorical variables. for example suppose we want to know
whether rentals that have an elevator are more likely to have a gym

```python
data.loc[(data["has_elevator"] ==0),"has_elevator"] ="No"
data.loc[(data["has_elevator"] ==1),"has_elevator"] ="Yes"
data.loc[(data["has_gym"] ==0),"has_gym"] ="No"
data.loc[(data["has_gym"] ==1),"has_gym"] ="Yes"

sns.countplot(x="has_elevator",hue="has_gym", data=data);
```

```python
#Sometimes, a data analyst is interested in simultaneously exploring the relationship between three or more variables in a single visualization. 
#Many of the visualization methods presented up to this point can include additional variables by using visual cues such as colors, shapes, and patterns.
#For example, we can investigate the relationship between rental price, square footage, and borough by using color to introduce our third variable:
plt.figure(figsize=(10,8))
sns.scatterplot(x='size_sqft', y='rent', hue='borough', data=data)
plt.show()
```
